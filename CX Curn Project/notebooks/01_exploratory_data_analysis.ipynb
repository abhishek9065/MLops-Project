````python
import sys
from pathlib import Path

# Make the project importable when running from the notebooks/ folder.
sys.path.append(str(Path.cwd().parent))

import pandas as pd
import matplotlib.pyplot as plt

from src.config import RAW_DATA_FILE, TARGET, NUMERIC_FEATURES, CATEGORICAL_FEATURES

df = pd.read_csv(RAW_DATA_FILE)
print(df.shape)
df.head()
````

## Target balance

Churn is imbalanced — this is why we track ROC-AUC / F1, not just accuracy.

````python
churn_rate = df[TARGET].mean()
print(f'Churn rate: {churn_rate:.1%}')
df[TARGET].value_counts().plot(kind='bar', title='Churn (1) vs Stay (0)')
plt.show()
````

## Numeric features

````python
df[NUMERIC_FEATURES].describe()
````

## Which categories churn most?

Churn rate by contract type — month-to-month customers should churn far more.

````python
(df.groupby('contract_type')[TARGET].mean()
   .sort_values(ascending=False)
   .plot(kind='barh', title='Churn rate by contract type'))
plt.xlabel('Churn rate')
plt.show()
````

````python
# Correlation of numeric features with churn.
df[NUMERIC_FEATURES + [TARGET]].corr()[TARGET].sort_values(ascending=False)
````
